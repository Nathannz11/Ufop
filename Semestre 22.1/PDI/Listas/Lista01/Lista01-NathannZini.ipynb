```python
import numpy as np
x = np.array([1,2,3,4,5,6])

print("Questao 1")
print("a) ", x[5])
print("b) ", x[1:4])
print("c) ", x[1:])
print("d) ", x[: -1])
print("e) ", x[5:1:-1])
print("f) ", x[[1,5,2,1,1]])
```

```python
A = np.array([1,2,3]).reshape((1,3))
B = np.array([3,2,1]).reshape((1,3))
M = np.array([[4,5,6],[6,5,3]])

print("Questao 3")
print("a) ", A + B)
print("b) ", A + M)
print("c) ", A.T)
print("d) ", A.T + B)
print("e) ", M - np.vstack([A,B]))
#print("f) ", np.vstack([A, B.T])) #dá erro pois o tamanho de colunas de B, quando transposta, não é o mesmo que o tamanho de colunas A
print("g) ", np.vstack([A, B]))
print("h) ", M - 3)
print("i)", A * B)
print("j)", np.matmul(A, B.T))
```
