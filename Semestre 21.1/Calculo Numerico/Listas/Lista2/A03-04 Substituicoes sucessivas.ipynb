#**Implementação do algoritmo de substituições sucessivas**

**Definição da função de substituição sucessiva.**
* A função recebe como parâmetro uma matriz (lista de lista) e um vetor
* Vale lembrar que a matriz é quadrada e que o vetor possui o mesmo numero de linhas da matriz

```
def substituicoesSucessivas(L,c): #L = matriz(nxn) c = vetor com n posicoes
	x=[]
	n=len(c) # numero de posicoes de c (que eh igual ao valor de n)
	for i in range(n):
		x.append(0)
	
	# Incicio do algoritmo de substituicoes sucessivas
	x[0] = c[0]/L[0][0] 
	
	for i in range(1,n):
		soma=0
		for j in range(i):
			soma+=L[i][j]*x[j]
		
		x[i]=(c[i]-soma)/L[i][i]
	return x # x possui o mesmo tamanho de c
	
  # Fim do algoritmo de substituicoes sucessivas
```

Definição da função principal (*main*): utilizada para executar a função implementa acima.

```
def main():
  # Matriz de exemplo dos slides
  A=[[2, 0, 0, 0], [3, 5, 0, 0], [1, -6, 8, 0], [1, 4, -3, 9]]
  b=[4, 1, 48, 6]
  resposta = substituicoesSucessivas(A,b)
  print(resposta)


main()
```
