### Código com passagem de parâmetros
<p>Implementação do filtro sigma</p>

```python
def sigma_block(block, T):
  total = np.prod(block.shape)
  pos = int(total/2)
  m_min = block[pos]-T
  m_max = block[pos]+T
  ind = np.logical_and(block >= m_min, block <= m_max)
  values = np.mean(block[ind])
  return values
```

```python
def filtro_sigma(img, sigma = 20, cell=(3,3)):
  lin, col = cell
  footprint = np.ones( (lin,col) )
  T = 2 * sigma
  nimg = ndimage.generic_filter(img_as_float(img), sigma_block, footprint=footprint, extra_arguments=(T,))
  return nimg
```

```python
nimg = filtro_sigma(lenna, sigma=20, cell=(7,7))
f,ax = plt.subplots(1,2,figsize=(16,16))
ax[0].imshow(lenna, cmap='gray')
ax[1].imshow(nimg, cmap='gray')
```
